```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.util import random_noise
```

```python
image = cv2.imread("/kaggle/input/accident-detection-from-cctv-footage/data/train/Accident/acc1 (13).jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
noisy_image = random_noise(image, mode='s&p')
noisy_image = (255 * noisy_image).astype(np.uint8)
```

```python
mean_filtered = cv2.blur(noisy_image, (5,5))
median_filtered = cv2.medianBlur(noisy_image, 5)
gaussian_filtered = cv2.GaussianBlur(noisy_image, (5,5), 0)
```

```python
titles = ['Asli', 'Noisy', 'Mean', 'Median', 'Gaussian']
images = [image, noisy_image, mean_filtered, median_filtered, gaussian_filtered]

plt.figure(figsize=(15,4))
for i in range(5):
    plt.subplot(1,5,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
